```python
# from google.colab import drive, files
from datetime import datetime as dt
import pandas as pd
import time
import warnings

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
# url link from - https://www.redfin.com/news/data-center/
redfin_url = 'https://redfin-public-data.s3.us-west-2.amazonaws.com/redfin_market_tracker/zip_code_market_tracker.tsv000.gz'


df = pd.read_csv(redfin_url, compression='gzip', sep='\t', on_bad_lines='warn')

print('Num of rows:', len(df))
print('Num of cols:', len(df.columns))
df.head()
```

```python
df.info()
```

```python
#let's change the period_begin and period_end to date time object and extract years and month.
df['period_begin'] = pd.to_datetime(df['period_begin'])
df['period_end'] = pd.to_datetime(df['period_end'])
df["period_begin_in_years"] = df['period_begin'].dt.year
df["period_end_in_years"] = df['period_end'].dt.year

df["period_begin_in_months"] = df['period_begin'].dt.month
df["period_end_in_months"] = df['period_end'].dt.month
```

```python
df.head(1)
```

```python
#let's map the month number to their respective month name.
month_dict = {
    "period_begin_in_months": {
        1 : "Jan",
         2 :  "Feb",
         3: "Mar",
         4: "Apr",
         5: "May",
         6: "Jun",
         7: "Jul",
         8: "Aug",
         9: "Sep",
         10: "Oct",
         11: "Nov",
         12: "Dec",
    },
    "period_end_in_months": {
        1 : "Jan",
         2 :  "Feb",
         3: "Mar",
         4: "Apr",
         5: "May",
         6: "Jun",
         7: "Jul",
         8: "Aug",
         9: "Sep",
         10: "Oct",
         11: "Nov",
         12: "Dec"
    }}

df = df.replace(month_dict)

df.head(2)
```

```python
#number of years in our dataset
sorted(df["period_begin_in_years"].unique())
```

Since most of the data in the city column are NAN. Let's get the city and county for our zip codes by joining our data with 
data obtained from https://raw.githubusercontent.com/scpike/us-state-county-zip/master/geo-data.csv

```python
# get zip code, city and county data
zip_county_url = 'https://raw.githubusercontent.com/scpike/us-state-county-zip/master/geo-data.csv'
zip_county_df = pd.read_csv(zip_county_url)
print('Num of rows:', len(zip_county_df))
print('Num of columns:', len(zip_county_df.columns))
zip_county_df.head()
```

```python
# check if zip code is valid by checking if the zip code is numeric.
zip_county_df['zip_code_isValid'] = zip_county_df.apply(lambda x: x['zipcode'].isnumeric(), axis=1)
print(zip_county_df.groupby(['zip_code_isValid', 'county'])['zipcode'].count())
```

```python
zip_county_df.head(1)
```

```python
#let's check for invalid zip codes to have an idea of how an invalid zip code looks.

print('Invalid zip codes:', zip_county_df.loc[zip_county_df['zip_code_isValid'] == False]['zipcode'].unique())
```

```python
# let's grab only the valid zip codes.
zip_county_df = zip_county_df.loc[zip_county_df['zip_code_isValid'] == True]
print('Length of valid zip code is', len(zip_county_df))
zip_county_df.head(1)
```

```python
# extract zip code from the region column from our redfin dataset
df['zipcode'] = df.apply(lambda x: x['region'].split(':')[1].strip(), axis=1)

# Now, let's merge the redfin dataset with our geo data
df_merge = pd.merge(
    df,
    zip_county_df[['zipcode', 'city', 'county']],
    how='inner', 
    on=['zipcode'] 
)

df_merge.head()
```

```python
df_merge.drop(["city_x", "region"], axis = 1, inplace=True)
df_merge = df_merge.rename(columns={'city_y': 'city'})
df_merge.head(1)
```
